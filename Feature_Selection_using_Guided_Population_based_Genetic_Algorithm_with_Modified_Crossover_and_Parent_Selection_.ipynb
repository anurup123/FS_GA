```python
import numpy as np
from sklearn.neighbors import KNeighborsClassifier as KNN
import matplotlib.pyplot as plt

class Solution():

    def __init__(self):
        self.num_features = None
        self.num_agents = None
        self.max_iter = None
        self.obj_function = None
        self.obj_function_ps=None
        self.execution_time = None
        self.convergence_curve = {}
        self.best_agent = None
        self.best_fitness = None
        self.best_accuracy = None
        self.final_population = None
        self.final_fitness = None
        self.final_accuracy = None


class Data():

    def __init__(self):
        self.train_X = None
        self.train_Y = None
        self.val_X = None
        self.val_Y = None



def initialize(num_agents, num_features):

    min_features = int(0.3 * num_features)
    max_features = int(0.6 * num_features)


    agents = np.zeros((num_agents, num_features))


    for agent_no in range(num_agents):


        cur_count = np.random.randint(min_features, max_features)
        temp_vec = np.random.rand(1, num_features)
        temp_idx = np.argsort(temp_vec)[0][0:cur_count]


        agents[agent_no][temp_idx] = 1

    return agents



def sort_agents(agents, obj, data, fitness=None):

    train_X, val_X, train_Y, val_Y = data.train_X, data.val_X, data.train_Y, data.val_Y
    (obj_function, weight_acc) = obj

    if fitness is None:

        if len(agents.shape) == 1:
            num_agents = 1
            fitness = obj_function(agents, train_X, val_X, train_Y, val_Y, weight_acc)
            return agents, fitness


        else:
            num_agents = agents.shape[0]
            fitness = np.zeros(num_agents)
            for id, agent in enumerate(agents):
                fitness[id] = obj_function(agent, train_X, val_X, train_Y, val_Y, weight_acc)

    idx = np.argsort(-fitness)
    sorted_agents = agents[idx].copy()
    sorted_fitness = fitness[idx].copy()

    return sorted_agents, sorted_fitness


def sort_agents_ps(agents, obj, data, fitness=None):

    train_X, val_X, train_Y, val_Y = data.train_X, data.val_X, data.train_Y, data.val_Y
    (obj_function_ps, weight_acc) = obj

    if fitness is None:

        if len(agents.shape) == 1:
            num_agents = 1
            fitness = obj_function_ps(agents, train_X, val_X, train_Y, val_Y, weight_acc)
            return agents, fitness


        else:
            num_agents = agents.shape[0]
            fitness = np.zeros(num_agents)
            for id, agent in enumerate(agents):
                fitness[id] = obj_function_ps(agent, train_X, val_X, train_Y, val_Y, weight_acc)

    idx = np.argsort(-fitness)
    sorted_agents = agents[idx].copy()
    sorted_fitness = fitness[idx].copy()

    return sorted_agents, sorted_fitness

def display(agents, fitness, agent_name='Agent'):

    print('\nNumber of agents: {}'.format(agents.shape[0]))
    print('\n------------- Best Agent ---------------')
    print('Fitness: {}'.format(fitness[0]))
    print('Number of Features: {}'.format(int(np.sum(agents[0]))))
    print('----------------------------------------\n')

    for id, agent in enumerate(agents):
        print('{} {} - Fitness: {}, Number of Features: {}'.format(agent_name, id+1, fitness[id], int(np.sum(agent))))

    print('================================================================================\n')



def compute_accuracy(agent, train_X, test_X, train_Y, test_Y):

    cols = np.flatnonzero(agent)
    if(cols.shape[0] == 0):
        return 0
    clf = KNN()

    train_data = train_X[:,cols]
    train_label = train_Y
    test_data = test_X[:,cols]
    test_label = test_Y

    clf.fit(train_data,train_label)
    acc = clf.score(test_data,test_label)

    return acc


def compute_fitness(agent, train_X, test_X, train_Y, test_Y, weight_acc=0.9):

    if(weight_acc == None):
        weight_acc = 0.9

    weight_feat = 1 - weight_acc
    num_features = agent.shape[0]

    acc = compute_accuracy(agent, train_X, test_X, train_Y, test_Y)
    feat = (num_features - np.sum(agent))/num_features

    fitness = weight_acc * acc + weight_feat * feat
    return fitness

def compute_fitness_ps(agent, train_X, test_X, train_Y, test_Y, weight_acc=0.9):

    if(weight_acc == None):
        weight_acc = 0.9

    weight_feat = 1 - weight_acc
    num_features = agent.shape[0]

    acc = compute_accuracy_(agent, train_X, test_X, train_Y, test_Y)
    feat = (num_features - np.sum(agent))/num_features

    fitness = weight_acc * acc + weight_feat * feat
    return fitness


def Conv_plot(convergence_curve):

    num_iter = len(convergence_curve['fitness'])
    iters = np.arange(num_iter) + 1
    fig, axes = plt.subplots(1)
    fig.tight_layout(pad = 5)
    fig.suptitle('Convergence Curves')

    axes.set_title('Convergence of Fitness over Iterations')
    axes.set_xlabel('Iteration')
    axes.set_ylabel('Avg. Fitness')
    axes.plot(iters, convergence_curve['fitness'])

    return fig, axes
```

```python
b=20
c=0.3
finalfeatures=[]
def mutate(agent):
        percent=0.2
        numChange=int(num_features*percent)
        pos=np.random.randint(0,num_features-1,numChange)
        agent[pos]=1-agent[pos]
        return agent
def GA(num_agents, max_iter, train_data, train_label, obj_function=compute_fitness, prob_cross=0.4, prob_mut=0.3, save_conv_graph=False, seed=0,obj_function_ps=compute_fitness_ps):
    # Genetic Algorithm
    ############################### Parameters ####################################
    #                                                                             #
    #   num_agents: number of chromosomes                                         #
    #   max_iter: maximum number of generations                                   #
    #   train_data: training samples of data                                      #
    #   train_label: class labels for the training samples                        #
    #   obj_function: the function to maximize while doing feature selection      #
    #   prob_cross: probability of crossover                                      #
    #   prob_mut: probability of mutation                                         #
    #   save_conv_graph: boolean value for saving convergence graph               #
    #                                                                             #
    ###############################################################################
    list_ = []

    short_name = 'GA'
    agent_name = 'Chromosome'
    train_data, train_label = np.array(train_data), np.array(train_label)
    num_features = train_data.shape[1]
    cross_limit = 5
    np.random.seed(seed)
    obj_function_ps=compute_fitness_ps
    mean_best=0

    weight_acc = None
    if(obj_function==compute_fitness):
        weight_acc = 0.95

    obj = (obj_function, weight_acc)

    obj_ = (obj_function_ps, weight_acc)
    compute_accuracy = (compute_fitness, 1)
    chromosomes=np.array(binary_population)
    fitness = np.zeros(num_agents)
    fitness_ps = np.zeros(num_agents)
    accuracy = np.zeros(num_agents)
    Leader_agent = np.zeros((1, num_features))
    Leader_fitness = float("-inf")
    Leader_accuracy = float("-inf")

    convergence_curve = {}
    convergence_curve['fitness'] = np.zeros(max_iter)

    data = Data()
    val_size = 0.2
    data.train_X, data.val_X, data.train_Y, data.val_Y = train_test_split(train_data, train_label, stratify=train_label, test_size=val_size)

    solution = Solution()
    solution.num_agents = num_agents
    solution.max_iter = max_iter
    solution.num_features = num_features
    solution.obj_function = obj_function
    solution.obj_function_ps=obj_function_ps

    chromosomes, fitness = sort_agents(chromosomes, obj, data)
    start_time = time.time()

    for iter_no in range(max_iter):
        print('\n================================================================================')
        print('                          Iteration - {}'.format(iter_no+1))
        print('================================================================================\n')

        chromosomes, fitness = cross_mut(chromosomes, fitness, obj, data, prob_cross, cross_limit, prob_mut)

        chromosomes, fitness = sort_agents(chromosomes, obj, data, fitness)
        display(chromosomes, fitness, agent_name)

        if fitness[0]>Leader_fitness:
            Leader_agent = chromosomes[0].copy()
            Leader_fitness = fitness[0].copy()
        if fitness[0]> mean_best:
            mean_best = fitness[0]

        convergence_curve['fitness'][iter_no] = mean_best
        list_.append(mean_best)
    Leader_agent, Leader_accuracy = sort_agents(Leader_agent, compute_accuracy, data)
    chromosomes, accuracy = sort_agents(chromosomes, compute_accuracy, data)

    print('\n================================================================================')
    print('                                    Final Result                                  ')
    print('================================================================================\n')
    print('Leader ' + agent_name + ' Dimension : {}'.format(int(np.sum(Leader_agent))))
    print(Leader_agent)
    print('Leader ' + agent_name + ' Fitness : {}'.format(Leader_fitness))
    print('Leader ' + agent_name + ' Classification Accuracy : {}'.format(Leader_accuracy))
    print('\n================================================================================\n')

    end_time = time.time()
    exec_time = end_time - start_time

    fig, axes = Conv_plot(convergence_curve)
    if(save_conv_graph):
        plt.savefig('convergence_graph_'+ short_name + '.jpg')
    plt.show()
    print(list_)
    solution.best_agent = Leader_agent
    solution.best_fitness = Leader_fitness
    finalfeatures=Leader_agent
    solution.best_accuracy = Leader_accuracy
    solution.convergence_curve = convergence_curve
    solution.final_population = chromosomes
    solution.final_fitness = fitness
    solution.final_accuracy = accuracy
    solution.execution_time = exec_time

    return solution


import numpy as np
import numpy as np
import numpy as np

def binary_to_decimal(binary_list):
    decimal_value = 0
    for bit in binary_list:
        decimal_value = (decimal_value << 1) | bit
    return int(decimal_value)



def horizontal_crossover(parent_1, parent_2):
    # print(parent_1)
    # print(parent_2)
    assert len(parent_1) == len(parent_2), "Both parents should have the same length."
    # print("hello")
    decimal_parent_1 = binary_to_decimal(parent_1)
    decimal_parent_2 = binary_to_decimal(parent_2)
    delta1, delta2 = np.random.rand(2)
    epsilon1, epsilon2 = np.random.uniform(-1, 1, 2)

    new_decimal_1 = delta1 * decimal_parent_1 + (1 - delta1) * decimal_parent_2 + epsilon1 * (decimal_parent_1 - decimal_parent_2)
    new_decimal_2 = delta2 * decimal_parent_2 + (1 - delta2) * decimal_parent_1 + epsilon2 * (decimal_parent_2 - decimal_parent_1)

    new_decimal_1 = int(new_decimal_1)
    new_decimal_2 = int(new_decimal_2)
    # print(new_decimal_1)
    # print(new_decimal_2)
    new_decimal_1 = np.clip(new_decimal_1, 0, 2**len(parent_1) - 1)
    new_decimal_2 = np.clip(new_decimal_2, 0, 2**len(parent_1) - 1)

    child_1 = [int(bit) for bit in np.binary_repr(int(new_decimal_1), width=len(parent_1))]
    child_2 = [int(bit) for bit in np.binary_repr(int(new_decimal_2), width=len(parent_1))]

    return child_1, child_2



def crossover(parent_1, parent_2, prob_cross):
    assert len(parent_1) == len(parent_2), "Both parents should have the same length."

    idx = np.arange(len(parent_1))
    np.random.shuffle(idx)

    shuffled_parent_1 = parent_1[idx]
    shuffled_parent_2 = parent_2[idx]

    child_1 = shuffled_parent_1.copy()
    child_2 = shuffled_parent_2.copy()

    for i in range(len(parent_1)):
        if np.random.rand() < prob_cross:
            child_1[i] = shuffled_parent_2[i]
            child_2[i] = shuffled_parent_1[i]

    child_1 = child_1[np.argsort(idx)]
    child_2 = child_2[np.argsort(idx)]

    return child_1, child_2

def mutation(chromosome, prob_mut):
    num_features = (np.array(chromosome)).shape[0]
    mut_chromosome = chromosome.copy()

    for i in range(num_features):
        if(np.random.rand()<prob_mut):
            mut_chromosome[i] = 1-mut_chromosome[i]

    return mut_chromosome


def roulette_wheel(fitness):
    maximum = sum([f for f in fitness])
    selection_probs = [f/maximum for f in fitness]
    return np.random.choice(len(fitness), p=selection_probs)



import numpy as np

def hamming_distance(chrom1, chrom2):
    """Calculate the Hamming distance between two chromosomes"""
    return sum(c1 != c2 for c1, c2 in zip(chrom1, chrom2))


import numpy as np

def adaptive_threshold_roulette_wheel(fitness, steepness=1, midpoint=0.5, diversity_boost=0.5):
    if len(fitness) == 0:
        raise ValueError("Fitness list cannot be empty.")
    if any(f < 0 for f in fitness):
        raise ValueError("Fitness scores must be non-negative.")

    if len(set(fitness)) == 1 :
        diversity_normalized = 0
    else:
        values, counts = np.unique(fitness, return_counts=True)
        probabilities = counts / counts.sum()
        diversity = -np.sum(probabilities * np.log2(probabilities + 1e-10))
        max_diversity = np.log2(len(fitness))
        diversity_normalized = diversity / max_diversity

    diversity_scaled = steepness * (diversity_normalized - midpoint)
    pressure_threshold = 1 / (1 + np.exp(diversity_scaled))
    adjusted_fitness = [f * pressure_threshold for f in fitness]
    base_probability = 1 / len(fitness)

    combined_fitness = [af + diversity_boost * base_probability for af in adjusted_fitness]
    if not any(combined_fitness):
        raise ValueError("Combined fitness scores are all zero, selection is not possible.")

    total_combined_fitness = sum(combined_fitness)
    selection_probs = [f / total_combined_fitness for f in combined_fitness]

    return np.random.choice(len(fitness), p=selection_probs)




def cross_mut(chromosomes, fitness, obj, data, prob_cross, cross_limit, prob_mut):
    count = 0
    num_agents = chromosomes.shape[0]
    print('Crossover-Mutation phase starting....')

    while(count<cross_limit):
        print('\nCrossover no. {}'.format(count+1))
        id_1=adaptive_threshold_roulette_wheel(fitness)
        id_2=adaptive_threshold_roulette_wheel(fitness)
        if(id_1 != id_2) :
            try :
              child_1, child_2 = horizontal_crossover(chromosomes[id_1], chromosomes[id_2])
            except Exception as e :
              print(e)
              print('Crossover failed....')
              continue
            child_1 = mutation(child_1, prob_mut)
            child_2 = mutation(child_2, prob_mut)

            child = np.array([child_1, child_2])
            child, child_fitness = sort_agents(child, obj, data)

            for i in range(2):
                for j in range(num_agents):
                    if(child_fitness[i] > fitness[j]):
                        print('child {} replaced with chromosome having id {}'.format(i+1, j+1))
                        chromosomes[j] = child[i]
                        fitness[j] = child_fitness[i]
                        break

            count = count+1
        else:
            print('Crossover failed....')
            print('Restarting crossover....\n')

    return chromosomes, fitness





!git clone https://github.com/ghoshsoumyajit7/AAPSO.git
#UCI
dataset='/content/AAPSO/data/UCIcsv/BreastCancer.csv'
df=pd.read_csv(dataset)
(a,b)=np.shape(df)
data = df.values[:,0:b-1]
label = df.values[:,b-1]
num_features=data.shape[1]
it=1
while it>0:
  GA(28, 60,data, label, save_conv_graph=True)
  it=it-1
```
